```python
import pandas as pd
import zipfile
import random
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

```python
mean = []
for i in range(1,10,1):
    sample = pd.Series(random.sample(list(df['temp_ann']), i))
    mean.append(sample.mean())

sns.lineplot(mean)
```

```python

sns.boxplot(
    data= df,
    x='temp_ann',
    y='Salmo trutta fario',
    orient="h",
    width=0.4,
    fliersize=1,
    whis= 1,
    color='#319ebc', # HEX color picked from here: https://coolors.co/palettes/trending - don't forget to start by "#"
    )
sns.stripplot(
    data= df,
    x='temp_ann',
    y='Salmo trutta fario',
    orient="h",
    jitter=0.2, # the ammount of jitter (i.e. random point dispersion along the x-axis).
    linewidth=0, # no line around the poits
    color='#9e2a3f', # HEX color picked from here: https://coolors.co/palettes/trending - don't forget to start by "#"
    alpha=.2,
    s=2, # point size
    )
plt.tick_params(left=False) # removes tickmarks
sns.despine(left=True)
plt.xlabel('Mean Annual Temperature Cº')
plt.ylabel(' yes fish =1                        no fish= 0')
```
