### Use the Univariate_analysis.ipynb file and the dataset in EFIplus_medit.zip for this exercise:

```python
import pandas as pd
import zipfile
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
```

```python
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
# print(df)
```

Let's check the four catchments with the highest number of fish sampling sites:

```python
df['Catchment_name'].value_counts()
```

**Ebro, Galiza-Norte, Minho and Tejo** are the four catchments with the highest number of fish sampling sites.

Now we only want to create graphs for those four catchments, so we will create 4 subsets from our dataframe (Ebro, Galiza-Norte, Minho and Tejo).

```python
ebro = df[df['Catchment_name']=='Ebro']
galiza_norte = df[df['Catchment_name']=='Galiza-Norte']
minho = df[df['Catchment_name']=='Minho']
tejo = df[df['Catchment_name']=='Tejo']
```

Let's plot strip-plots, histograms and boxplots of Annual Mean Temperature (temp_ann) for Ebro, Galiza-Norte, Minho and Tejo.

### Strip-plot

```python
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

ax=plt.subplot(2,2,1)
ax.set(ylabel='Annual Mean Temperature')
sns.stripplot(data=ebro, y='temp_ann').set_title("Ebro")

ay=plt.subplot(2,2,2)
ay.set(ylabel='Annual Mean Temperature')
sns.stripplot(data=galiza_norte, y='temp_ann').set_title("Galiza-Norte")

aw=plt.subplot(2,2,3)
aw.set(ylabel='Annual Mean Temperature')
sns.stripplot(data=minho, y='temp_ann').set_title("Minho")

az=plt.subplot(2,2,4)
az.set(ylabel='Annual Mean Temperature')
sns.stripplot(data=tejo, y='temp_ann').set_title("Tejo")

plt.show()
```

### Histogram

```python
plt.subplots_adjust(left=0.1,
                    bottom=0.01,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.7)

plt.subplot(2,2,1)
sns.histplot(
    ebro["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Ebro")

plt.subplot(2,2,2)
sns.histplot(
    galiza_norte["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Galiza-Norte")

plt.subplot(2,2,3)
sns.histplot(
    minho["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Minho")

plt.subplot(2,2,4)
sns.histplot(
    tejo["temp_ann"], 
    kde=True,
    stat="density",
    kde_kws=dict(cut=3),
    alpha=.4,
    edgecolor=(1, 1, 1, 0.4),
).set_title("Tejo")

plt.show()
```

### Boxplot

```python
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.subplot(2,2,1)
sns.boxplot(data=ebro, y='temp_ann').set_title("Ebro")

plt.subplot(2,2,2)
sns.boxplot(data=galiza_norte, y='temp_ann').set_title("Galiza-Norte")

plt.subplot(2,2,3)
sns.boxplot(data=minho, y='temp_ann').set_title("Minho")

plt.subplot(2,2,4)
sns.boxplot(data=tejo, y='temp_ann').set_title("Tejo")

plt.show()
```
