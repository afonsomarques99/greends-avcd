```python
import pandas as pd
import zipfile
import seaborn as sns # For plotting
import matplotlib.pyplot as plt #For showing plots
```

```python
df = pd.read_csv('EFIplus_medit.zip',compression='zip', sep=";")
```

```python
#list(df.columns)
```

```python
catchment_count = pd.crosstab(index = df['Catchment_name'], columns='count')
#print(catchment_count)
```

```python
catchment_4_higher_fish = catchment_count["count"].nlargest(4)
print(catchment_4_higher_fish)
```

```python
catchment_Ebro = df[df['Catchment_name'] == 'Ebro'] #Selecciona as linhas da Dataframe nas quais o valor da coluna 'Catchment_name' e igual a 'Ebro'
#print(catchment_Ebro)
#type(catchment_Ebro)

catchment_Galiza_Norte = df[df['Catchment_name'] == 'Galiza-Norte']

catchment_Minho = df[df['Catchment_name'] == 'Minho']

catchment_Tejo = df[df['Catchment_name'] == 'Tejo']
```

```python
# EBRO CATCHMENT
# Create a figure with 4 subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Set Seaborn style
sns.set_style("whitegrid")

# Histogram
sns.histplot(ax=axes[0, 0], data=catchment_Ebro, x="temp_ann", color = 'orange')

# Violin plot
sns.violinplot(ax=axes[0, 1], data=catchment_Ebro, x="Catchment_name", y="temp_ann", color = 'yellow')

# Box plot 
sns.boxplot(ax=axes[1, 0], data=catchment_Ebro, x="Catchment_name", y="temp_ann", color = 'green')

# Strip plot 

sns.stripplot(ax=axes[1, 1], data=catchment_Ebro, x="Catchment_name", y="temp_ann", jitter=True, color = 'blue')

# Set title 
fig.suptitle("Annual Mean Temperature in Ebro Catchment")

# Show the plot
plt.show()
```
